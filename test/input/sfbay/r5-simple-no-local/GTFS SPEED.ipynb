```python
# Import package
import gtfs_functions as gtfs
import pandas as pd
from pandas import DataFrame
import numpy as np
```

```python
filenames = ['3D.zip', 'AC.zip', 'AM.zip', 'AY.zip', 'BA.zip', 'Caltrain.zip', 'CC.zip', 'CE.zip', 'CM.zip',
            'CT.zip', 'DE.zip', 'EM.zip', 'GG.zip', 'HF.zip', 'MA.zip', 'PE.zip', 'RV.zip', 'SB.zip', 'SC.zip',
            'SF.zip', 'SM.zip', 'SO.zip', 'SR.zip', 'ST.zip', 'TD.zip', 'UC.zip', 'VC.zip', 'VN.zip', 'VT.zip',
            'WC.zip', 'WH.zip']
names = ['3D', 'AC', 'AM', 'AY', 'BA', 'Caltrain', 'CC', 'CE', 'CM',
            'CT', 'DE', 'EM', 'GG', 'HF', 'MA', 'PE', 'RV', 'SB', 'SC',
            'SF', 'SM', 'SO', 'SR', 'ST', 'TD', 'UC', 'VC', 'VN', 'VT',
            'WC', 'WH']
```

```python
for filename in filenames:
    try:  
        routes, stops, stop_times, trips, shapes = gtfs.import_gtfs(filename)
        segments_gdf = gtfs.cut_gtfs(stop_times, stops, shapes)
        cutoffs = cutoffs = list(range(24))
        speeds = gtfs.speeds_from_gtfs(routes, stop_times, segments_gdf, cutoffs = cutoffs)
        speeds['agency'] = filename
        speeds.to_csv(filename + 'SPEED.csv')
    except:
        print(filename)
        continue
```

```python
Aggregation = DataFrame()
for filename, name in zip(filenames, names):
    try:  
        routes, stops, stop_times, trips, shapes = gtfs.import_gtfs(filename)
        segments_gdf = gtfs.cut_gtfs(stop_times, stops, shapes)
        cutoffs = cutoffs = list(range(24))
        speeds = gtfs.speeds_from_gtfs(routes, stop_times, segments_gdf, cutoffs = cutoffs)
        speeds['agency'] = name
        Aggregation = Aggregation.append(speeds)
    except:
        print(filename)
        continue
```
